## Problem 1

Write a function called `group_means` that takes its input a Pandas DataFrame with exactly the columns: "labels" and "weights". Group entries by labels and return the average weights for each label. Make sure to name the column in the returned data frame "average weight" and use the "labels" as the index, but don't name the index column. 

**Example:** See an example by executing the code in the second cell below.

```python
import pandas as pd
import numpy as np
from IPython.display import display

def group_means(X):
    # YOUR CODE HERE
    Y = X.groupby(["labels"]).mean() #group the data by label and take the mean of each group
    Y.rename(columns = {"weights": "average weight"}, inplace = True) #rename weights to average weight
    Y.index.name = None #remove the index name
    return Y

    
    
```

    * income group of the country.
    
Display your table in Data Frame format in your notebook. 

```python
from pathlib import Path 
import pandas as pd

Path.cwd()

# YOUR CODE HERE
file_path1 = "C:/Users/Jen/Documents/Math 167PS/Homework/Homework06/WorldBank_Metadata_Countries.csv"
file_path2 = "C:/Users/Jen/Documents/Math 167PS/Homework/Homework06/WorldBank_Population.csv"  #store path with to data file
metadata = pd.read_csv(file_path1) #read to python
population = pd.read_csv(file_path2) #read to python

combined_data = pd.merge(metadata, population) #merge datasets

#1. include only the entries whose region is only in europe and central asia 
combined_data.set_index(["Region"], inplace = True) #sort by region
combined_data = combined_data.loc["Europe & Central Asia"] #return entries whose region is in Europe & Central Asia

#2. Use the "Country Name" as the row identifier or index value 
combined_data.set_index(["Country Name"], inplace= True) #index by Country Name

#3. Include the follow columns for each country: population count in 1988/2018, percent growth in population in %, income group
combined_data = combined_data[["1988", "2018", "IncomeGroup"]] #pick a subset of desired columns
combined_data.rename(columns = {"IncomeGroup":"Income Group"}, inplace = True) #rename Income Group for style purposes

#calculate all of the percentages for population growth
before = combined_data["1988"]
after = combined_data["2018"]
combined_data["Percent Growth in Population"] = round(((after-before)/before)*100,2)

final_data = combined_data[["1988", "2018", "Percent Growth in Population", "Income Group"]] #reorder the columns

final_data
```
