````python
import numpy as np

def mystery_1(array):
    """the input should be a two-dimensional NumPy array"""  
    statement = False
    foo = np.isnan(array)
    twix = []
    if array.ndim !=2:
        statement = print("wrong dimension")
    else:
        for bar in foo:
            twix.append(sum(bar))
        if sum(twix) == 0:
            statement = print("there are none")
        else: 
            statement = twix
    return statement
````

The function mystery_1 checks if the two-dimensional NumPy array contains NaN values in its rows. If there are no NaN values, the output is "there are none." Otherwise, the output returns a list whose entries correspond to the sum of NaN values in each row. 

````python
import numpy as np

def mystery_2(array):
    """the input should be a two-dimensional NumPy array"""  
    if array.ndim==2:
        j = array.shape[0]
        g = array.shape[1]
        foo = np.nansum(array, axis=1)/(np.sum(~np.isnan(array), axis = 1))
        for g in range(j):
            bar = np.isnan(array[g,:])
            array[g,bar] = foo[g]
        return array
````

The function mystery_2 replaces NaN values in the ith row by the sum over the ith row (assuming that NaN values are treated as 0) divided by the sum of the corresponding ith row of the inverted boolean array. The boolean array is determined by how many NaN values in the array, and the function takes the inversion of it.
